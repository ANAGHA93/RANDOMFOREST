```python


import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets  
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import  DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import classification_report
from sklearn import preprocessing
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import KFold

from sklearn.model_selection import cross_val_score
```

```python
#Read the file
df=pd.read_csv('Fraud_check.csv')
```

```python
df
```

```python
df=df.rename(columns={'Marital.Status':'MI','Taxable.Income':'TaxableIncome','Work.Experience':'WorkExperience'})
```

```python
#oneHotEncoding for the categorical input variables
enc = OneHotEncoder(handle_unknown='ignore')
```

```python
enc_df1 = pd.DataFrame(enc.fit_transform(df[['MI']]).toarray())
```

```python
x=enc_df1.rename(columns={0:'MI_DIV',1:'MI_MAR',2:'MI_SING'})
```

```python
x.columns
```

```python
x
```

```python
enc_df2 = pd.DataFrame(enc.fit_transform(df[['Undergrad']]).toarray())
```

```python
y=enc_df2.rename(columns={0:'ntUndrgrad',1:'Undergrad'})
```

```python
y
```

```python
enc_df3 = pd.DataFrame(enc.fit_transform(df[['Urban']]).toarray())
```

```python
z=enc_df3.rename(columns={0:'ntUrban',1:'Urban'})
```

```python
z
```

```python
data_final = df.iloc[:,0:6].join(x)
data_final
```

```python
data_final=data_final.drop(['Urban'],axis=1)
```

```python
data_final1 = data_final.iloc[:,2:9].join(y)
data_final1
```

```python
data_final2= data_final1.iloc[:,0:8].join(z)
data_final2
```

```python
#finding the maximum value of TaxableIncome
tax_max=data_final2['TaxableIncome'].max(axis=0)
tax_max
```

```python
#finding the minimum value of TaxableIncome
tax_min=data_final2['TaxableIncome'].min(axis=0)
tax_min
```

```python
#The output variable Taxgroup created and contents are made by splitting the TaxableIncome variable where TaxableIncome between 10003 and30000 are categorised as  "Risky" and between 30000 and 99619 as "Good"
data_final2['Taxgroup']=pd.cut(data_final2.TaxableIncome,bins=[10003,30000,99619],labels=["Risky","Good"])
data_final2
```

```python
#Label encoding done on output variable
label_encoder = preprocessing.LabelEncoder()
data_final2['Taxgroup']= label_encoder.fit_transform(data_final2['Taxgroup'])
```

```python
#Separating the input and output variables
x=data_final2.iloc[:,0:10]
y=data_final2["Taxgroup"]
```

```python
#defining the K folds
kfold=KFold(n_splits=10,random_state=7,shuffle=True)
```

```python
#defining the RandomForestClassifier and building the model
model=RandomForestClassifier(n_estimators=100,max_features=3)
model
```

```python
results=cross_val_score(model,x,y,cv=kfold)
```

```python
#accuracy of the model
print(results.mean())
```
